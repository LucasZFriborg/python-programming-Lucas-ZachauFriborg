<img src="../assets/numpy_1b.png" width = 300>

````python
import numpy as np
import matplotlib.pyplot as plt

# a)
theoretical_mean = np.mean([1, 2, 3, 4, 5, 6])

print(f'Theoretical mean of a six-sided dice: {theoretical_mean}')

# b)
throws = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
results = []

for throw in throws:
    result = np.random.randint(1, 7, size=throw)
    mean = np.mean(result)
    results.append(mean)

x = [i+1 for i in range(len(results))]

plt.title('Mean of a six-sided die')
plt.xlabel('Number of rolls')
plt.ylabel('Sample means')
plt.plot(x, results)
plt.show()
````
